The standarized regression model

$\frac{y-\bar{y}}{\sigma_y}=\sum_{i=1}^M\tilde{\beta_i}\frac{x_i-\bar{x}_i}{\sigma_{x_i}}+\tilde{\epsilon}$

is equivalent to the regression model

$y=\beta_0+\sum_{i=1}^M\beta_ix_i+\epsilon$

with

$\beta_0=\bar{y}-\sum_{i=1}^M\tilde{\beta_i}\frac{\sigma_i}{\sigma_{x_i}}\bar{x}_i$,

$\beta_i=\frac{\sigma_i}{\sigma_{x_i}}\tilde{\beta_i}\quad 1\le i\le M$,

$\epsilon=\sigma_y\tilde{\epsilon}$.

Can you prove it?

Below we regularize a scaled linear regression model. The regularization appears to reduce a little the noise of the predictions.

Import libraries and load the data

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from sklearn.linear_model import LinearRegression
import scipy.io

matplotlib.rcParams['lines.linewidth'] = 2
matplotlib.rcParams['axes.spines.right'] = False
matplotlib.rcParams['axes.spines.top'] = False


d = scipy.io.loadmat('joystick_track/data/rh_joystick.mat')

X = d['data']
y_raw = d['CursorPosY']

N = X.shape[0]
sample_freq = 1000
tmax = N/sample_freq
ts = np.linspace(0,tmax,N)

print(f"Data dimension: {X.shape}")
```

Writing our own linear regression function!

```python
def linreg(X, y, reg_coef=0.05, standarize=False, add_col_ones=True):
    assert X.ndim == 2
    assert X.shape[0] == y.shape[0]

    if standarize:
        y = (y - np.mean(y)) / np.std(y)
        X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    elif add_col_ones:
        X = np.hstack((X, np.ones((X.shape[0], 1))))

    Id = np.eye(X.shape[1])
    C = np.dot(X.T,X)+reg_coef*Id        # [P N] x [N P] = [P P]
    u = np.dot(X.T,y)        # [P N] x [N]   = [P]
    w = np.linalg.solve(C,u) # w=C^{-1}*u
    return w

def predict(X, w, standarize=False, add_col_ones=True):
    if standarize:
        X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    elif add_col_ones:
        X = np.hstack((X, np.ones((X.shape[0], 1))))

    predictions = np.dot(X, w)
    return predictions
```

Some processing:
- Adding a constant predictor to the design matrix (X)
- Discarding the first 10 seconds of the experiment
- standardizing the response variables.

```python
idx = ts>=10

X = X[idx,:]
ts = ts[idx]
y = y_raw[idx]

print(X.shape)
```

Fit the model on ~half the data, compute the prediction and residuals:

```python
reg_coef = 1e4
w = linreg(X[ts<=60,:],y[ts<=60], reg_coef=reg_coef, standarize=True, add_col_ones=False)
yhat_scaled = predict(X[ts<=60], w, standarize=True, add_col_ones=False)
yhat = yhat_scaled * np.std(y[ts<=60]) + np.mean(y[ts<=60])
err = y[ts<=60] - yhat
```

```python
plt.figure(figsize=(12,3))
plt.plot(ts[ts<=60],y[ts<=60],color='blue',linewidth=2.5, label="True")
plt.plot(ts[ts<=60],yhat,color='gray',alpha=0.6, label="Predicted")
plt.xlabel("Time (s)")
plt.ylabel("Cursor Y pos")
plt.legend(ncol=2,frameon=False,loc='lower left',fontsize='small')
```

```python
plt.hist(err,bins=300,density=True)
m = np.mean(err)
s = np.std(err)

xs_e = np.linspace(-3,3,2000)
gaussian_fit = (1/np.sqrt(2*np.pi*s**2))*np.exp(-0.5*((xs_e-m)/s)**2)
plt.plot(xs_e,gaussian_fit,color='red',linewidth=3)
plt.xlabel("Residuals")
plt.ylabel("Density")
```

Evaluate **the same model** on the rest of the data:

```python
yhat_scaled = predict(X[ts>60], w, standarize=True, add_col_ones=False)
yhat = yhat_scaled * np.std(y[ts>60]) + np.mean(y[ts>60])
err = y[ts>60] - yhat
nmse = np.mean(err**2)/np.var(y[ts>60])

plt.figure(figsize=(12,3))
plt.plot(ts[ts>60],y[ts>60],color='blue',linewidth=2.5, label="True")
plt.plot(ts[ts>60],yhat,color='gray',alpha=0.6, label="Predicted")
plt.title(f"NMSE={nmse}")
plt.xlabel("Time (s)")
plt.ylabel("Cursor Y pos")
plt.legend(ncol=2,frameon=False,loc='lower left',fontsize='small')
```

```python
plt.hist(err,bins=300,density=True)
m = np.mean(err)
s = np.std(err)

xs_e = np.linspace(-5,5,2000)
gaussian_fit = (1/np.sqrt(2*np.pi*s**2))*np.exp(-0.5*((xs_e-m)/s)**2)
plt.plot(xs_e,gaussian_fit,color='red',linewidth=3)
plt.xlabel("Residuals")
plt.ylabel("Density")
```
